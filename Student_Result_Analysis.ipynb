### Drop unnamed column

```python
df = df.drop("Unnamed: 0", axis=1)
print(df.head())
```

### Gender Distribution

```python
mtp.figure(figsize = (5,5))
ax= sns.countplot(data = df, x = "Gender")
mtp.title("Gender Distribution")
ax.bar_label(ax.containers[0])
mtp.show()
```

#### from the above chart we have analysed that the number of females in the data is more than number of males

```python
gb =  df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean", "WritingScore":"mean"})
print(gb)
```

```python
mtp.figure(figsize = (5,4))
sns.heatmap(gb, annot= True)
mtp.title("Relationship between parent education and student score")
mtp.show()
```

#### from the above chart we have concluded that the education of parents have a good impact on their scores

```python
gb1 =  df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean", "WritingScore":"mean"})
print(gb1)
```

```python
mtp.figure(figsize = (5,4))
sns.heatmap(gb1, annot= True)
mtp.title("Relationship between parent marital status and student score")
mtp.show()
```

#### from the above chart we have analysed that the marital status of parent has negligable impact on the student scores.

### Distribution of Ethnic Groups

```python
groupA= df.loc[(df["EthnicGroup"] == "group A")].count()
groupB= df.loc[(df["EthnicGroup"] == "group B")].count()
groupC= df.loc[(df["EthnicGroup"] == "group C")].count()
groupD= df.loc[(df["EthnicGroup"] == "group D")].count()
groupE= df.loc[(df["EthnicGroup"] == "group E")].count()

l=["group A", "group B","group C", "group D", "group E"]
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
mtp.pie(mlist, labels= l, autopct= "%1.2f%%")
mtp.title("Distribution of Ethnic groups")
mtp.show()
```
